```
# from random import choice
from datetime import datetime
import math

board = ["-", "-", "-",
         "-", "-", "-",
         "-", "-", "-"]
initial_depth = 0
is_game_in_process = True
valid_inputs = ["1", "2", "3", "4", "5", "6", "7", "8", "9"]

player = "O"
wins_x = 0
wins_o = 0
ties = 0


def display_board():
    print(board[0] + " | " + board[1] + " | " + board[2])
    print(board[3] + " | " + board[4] + " | " + board[5])
    print(board[6] + " | " + board[7] + " | " + board[8])


def best_move():
    global is_game_in_process
    slots = options_left()
    # print(slots)
    best_score = -math.inf
    position = 0
    for slot in slots:
        index = int(slot) - 1
        if board[index] == "-":
            board[index] = "X"
            score = minimax(board, 0, False)
            board[index] = "-"
            if score > best_score:
                best_score = score
                position = index
    is_game_in_process = True
    return position


scores = {
    "X": 10,
    "O": -10,
    "tie": 0
}


def minimax(board, depth, is_max):
    result = check_for_result()
    if result:
        return scores[result]
    if is_max:
        best_score = -math.inf
        slots = options_left()
        for slot in slots:
            index = int(slot) - 1
            board[index] = "X"
            score = minimax(board, depth+1, False)
            board[index] = "-"
            best_score = max(score, best_score)
        return best_score
    else:
        best_score = math.inf
        slots = options_left()
        for slot in slots:
            index = int(slot) - 1
            board[index] = "O"
            score = minimax(board, depth+1, True)
            board[index] = "-"
            best_score = min(score, best_score)
        return best_score


def options_left():
    options_available = [(x + 1) for x, letter in enumerate(board) if letter == '-']
    i = 0
    while i < len(options_available):
        options_available[i] = str(options_available[i])
        i = i + 1
    return options_available


def user_input():
    # options_available = options_left()
    global initial_depth
    if player == "X":
        position = best_move()
        # print(position)
        is_position_valid(str(position + 1))
    else:
        position = input("please select from following (1-9): ")
        is_position_valid(position)


def is_position_valid(position):
    if position not in valid_inputs:
        print("it's not a valid input ")
        user_input()
    else:
        # print("it's a valid input")
        position = int(position) - 1
        is_space_empty(position)


def is_space_empty(position):
    if board[position] == "-":
        board[position] = player
    else:
        user_input()


def who(x):
    global wins_x
    global wins_o
    if x == "X":
        wins_x = wins_x + 1
    elif x == "O":
        wins_o = wins_o + 1


def tied():
    global ties
    ties = ties + 1


def check_for_result():
    global is_game_in_process
    row_winner = check_rows()
    column_winner = check_column()
    diagonal_winner = check_diagonal()
    match_tie = if_tie()
    if row_winner or column_winner or diagonal_winner or match_tie:
        is_game_in_process = False
        if row_winner:
            who(row_winner)
            return row_winner
        elif column_winner:
            return column_winner
            # print('winner is ' + column_winner)
            who(column_winner)
        elif diagonal_winner:
            who(diagonal_winner)
            return diagonal_winner
        elif match_tie:
            # print('match ended in no result. it\'s a tie')
            tied()
            return "tie"


def check_rows():
    row_1 = board[0] == board[1] == board[2] != "-"
    row_2 = board[3] == board[4] == board[5] != "-"
    row_3 = board[6] == board[7] == board[8] != "-"
    if row_1:
        return board[0]
    elif row_2:
        return board[3]
    elif row_3:
        return board[6]


def check_column():
    global is_game_in_process
    column_1 = board[0] == board[3] == board[6] != "-"
    column_2 = board[1] == board[4] == board[7] != "-"
    column_3 = board[2] == board[5] == board[8] != "-"
    if column_1:
        return board[0]
    elif column_2:
        return board[1]
    elif column_3:
        return board[2]


def check_diagonal():
    global is_game_in_process
    diagonal_1 = board[0] == board[4] == board[8] != "-"
    diagonal_2 = board[2] == board[4] == board[6] != "-"
    if diagonal_1:
        return board[0]
    elif diagonal_2:
        return board[2]


def if_tie():
    if "-" not in board:
        return True


def flip_turn():
    global player
    if player == "X":
        player = "O"
    else:
        player = "X"


x = 0
start_game = datetime.now()
a = start_game.strftime("%M:%S")
while x < 1:
    x = x + 1
    is_game_in_process = True
    board = ["-", "-", "-",
             "-", "-", "-",
             "-", "-", "-"]
    while is_game_in_process:
        # global is_game_in_process
        display_board()
        print('it\'s your turn ' + player)
        user_input()
        check_for_result()
        flip_turn()

display_board()
winner = check_for_result()
if winner != "tie":
    print(winner + " won")
else:
    print("it ended in a " + winner)
end_game = datetime.now()
b = end_game.strftime("%M:%S")
print(end_game - start_game)
```
